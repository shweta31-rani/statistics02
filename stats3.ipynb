12.What is t-test and how it is used in hypothesis testing.
-f


Ans:A t-test is a statistical hypothesis test used to determine if there's a significant difference between the means of two groups. It's commonly used in hypothesis testing to assess if a treatment or intervention has a significant effect, or if two groups differ from each other.
How it's used in hypothesis testing:

   1. Formulating Hypotheses:
  A hypothesis is a statement about the population mean(s) that you want to test. For example, you might hypothesize that a new drug is more effective than a placebo (null hypothesis: the drug has no effect; alternative hypothesis: the drug has an effect).

2. Setting Significance Level:
A significance level (alpha) is chosen, typically 0.05, which determines the probability of rejecting the null hypothesis when it's actually true (Type I error).
3. Calculating the t-statistic:
The t-statistic is calculated using a formula that considers the differences between the sample means, the sample variances, and the sample sizes.
4. Comparing to the Critical Value or p-value:
The calculated t-statistic is compared to a critical value or a p-value. If the t-statistic exceeds the critical value or the p-value is less than the significance level, the null hypothesis is rejected.
5. Interpreting the Results:
Based on the comparison, you either reject or fail to reject the null hypothesis. If rejected, it suggests that the observed difference between the groups is unlikely to be due to chance.

18.What is chi-square goodness of fit test and how si it applied-f

Ans:he chi-square goodness of fit test is a statistical method used to determine how well a sample's distribution matches a theoretical distribution or population distribution.-f



Here's how it's applied:

 1. Formulating a Hypothesis:
You start with a null hypothesis that the sample data comes from a specific distribution (e.g., normal, binomial, uniform).

2. Calculating Expected Frequencies:
Based on the assumed distribution, you calculate the expected frequency for each category in your sample.
3. Calculating the Chi-Square Statistic:
You compare the observed frequencies with the expected frequencies using a chi-square statistic. This statistic measures the discrepancy between the observed and expected values.
4. Determining the p-value:
The chi-square statistic is compared to a chi-square distribution with appropriate degrees of freedom to determine the p-value.
5. Making a Decision:
If the p-value is below a chosen significance level (e.g., 0.05), you reject the null hypothesis and conclude that the sample distribution is significantly different from the assumed distribution

```python
def bernoulli_pdf(x, p):
    """
    Calculate the Probability Distribution Function (PDF) for a Bernoulli distribution.

    Parameters:
    x (int): The outcome for which the PDF is calculated (either 0 or 1).
    p (float): The probability of success (P(X = 1)).

    Returns:
    float: The probability of observing the outcome x.
    """
    if x == 1:
        return p
    elif x == 0:
        return 1 - p
    else:
        raise ValueError("Input x must be either 0 or 1.")

# Example usage:
p = 0.6  # Probability of success
x_values = [0, 1]

# Calculate PDF for both outcomes (x = 0 and x = 1)
pdf_values = [bernoulli_pdf(x, p) for x in x_values]

print(f"PDF for X = 0: {pdf_values[0]}")
print(f"PDF for X = 1: {pdf_values[1]}")
```

```python
#cdf
import numpy as np
import matplotlib.pyplot as plt

# Parameters of the discrete uniform distribution
a = 1   # minimum value
b = 6   # maximum value (inclusive)

# Support range (x-values for plotting)
x_vals = np.arange(a - 2, b + 3)  # Extend range to show tails

# CDF values
cdf_vals = []
for x in x_vals:
    if x < a:
        cdf = 0
    elif x > b:
        cdf = 1
    else:
        cdf = (np.floor(x) - a + 1) / (b - a + 1)
    cdf_vals.append(cdf)

# Plotting the CDF
plt.step(x_vals, cdf_vals, where='post', label='CDF', color='blue')
plt.scatter(x_vals, cdf_vals, color='blue')
plt.title(f"CDF of Discrete Uniform Distribution (a={a}, b={b})")
plt.xlabel('x')
plt.ylabel('F(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
from scipy.special import comb

# Parameters
n = 10      # number of trials
p = 0.5     # probability of success

# Step 1: Generate random binomial data
num_samples = 1000
data = np.random.binomial(n=n, p=p, size=num_samples)

# Step 2: Calculate frequencies of each outcome
unique_vals, counts = np.unique(data, return_counts=True)

# Step 3: Calculate exact binomial probabilities using the binomial formula
probabilities = [comb(n, k) * (p ** k) * ((1 - p) ** (n - k)) for k in unique_vals]

# Alternatively, using scipy.stats.binom.pmf:
# probabilities = binom.pmf(unique_vals, n, p)

# Step 4: Plot the results
plt.figure(figsize=(10, 5))
plt.bar(unique_vals, counts / num_samples, width=0.6, alpha=0.7, label='Simulated Frequencies')
plt.plot(unique_vals, probabilities, 'ro-', label='Theoretical Probabilities (PMF)')
plt.title(f'Binomial Distribution: n={n}, p={p}')
plt.xlabel('Number of Successes (k)')
plt.ylabel('Probability')
plt.xticks(range(n+1))
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# ---- Step 1: Input your data ----
x = 75           # data point
mean = 70        # population/sample mean
std = 5          # population/sample standard deviation

# ---- Step 2: Calculate z-score ----
z = (x - mean) / std
print(f"Z-score: {z:.2f}")

# ---- Step 3: Calculate probability ----
p_less_than_x = norm.cdf(z)
p_greater_than_x = 1 - p_less_than_x
p_two_tailed = 2 * min(p_less_than_x, p_greater_than_x)

print(f"P(Z < {z:.2f}) = {p_less_than_x:.4f}")
print(f"P(Z > {z:.2f}) = {p_greater_than_x:.4f}")
print(f"Two-tailed p-value = {p_two_tailed:.4f}")

# Interpret
if abs(z) > 2:
    print("This is an unusual value (|z| > 2).")
else:
    print("This is a usual value (|z| <= 2).")

# ---- Step 4: Plot the standard normal curve and mark z ----
x_vals = np.linspace(-4, 4, 1000)
y_vals = norm.pdf(x_vals)

plt.plot(x_vals, y_vals, label='Standard Normal PDF')
plt.fill_between(x_vals, y_vals, where=(x_vals <= z), color='skyblue', alpha=0.5, label=f'P(Z ≤ {z:.2f})')
plt.axvline(z, color='red', linestyle='--', label=f'z = {z:.2f}')
plt.title('Z-Score and Standard Normal Distribution')
plt.xlabel('Z')
plt.ylabel('Density')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.stats import norm

# Step 1: Sample data
sample = np.array([108, 112, 105, 98, 110, 101, 100, 102, 107, 106])
mu_0 = 100       # Hypothesized population mean
sigma = 15       # Known population standard deviation
alpha = 0.05     # Significance level

# Step 2: Calculate sample statistics
sample_mean = np.mean(sample)
n = len(sample)

# Step 3: Calculate z-statistic
z = (sample_mean - mu_0) / (sigma / np.sqrt(n))

# Step 4: Calculate two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z)))

# Step 5: Output results
print(f"Sample Mean = {sample_mean:.2f}")
print(f"Z-statistic = {z:.3f}")
print(f"P-value = {p_value:.4f}")

# Step 6: Decision
if p_value < alpha:
    print("Reject the null hypothesis: significant difference.")
else:
    print("Fail to reject the null hypothesis: no significant difference.")
```

```python
import numpy as np
import scipy.stats as stats

# Sample data
data = np.array([108, 112, 105, 98, 110, 101, 100, 102, 107, 106])

# Step 1: Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Use ddof=1 for sample standard deviation
n = len(data)

# Step 2: Calculate standard error (SE)
SE = sample_std / np.sqrt(n)

# Step 3: Find the critical value (z or t) for a 95% confidence level
confidence_level = 0.95
alpha = 1 - confidence_level
t_critical = stats.t.ppf(1 - alpha / 2, df=n - 1)  # t-distribution critical value (for small samples)

# Step 4: Calculate the margin of error
margin_of_error = t_critical * SE

# Step 5: Compute the confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Output results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"Standard Error: {SE:.2f}")
print(f"Critical Value (t): {t_critical:.3f}")
print(f"Margin of Error: {margin_of_error:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the parameters
population_mean = 50  # Population mean (H₀: µ = 50)
population_std = 10   # Population standard deviation
sample_size = 30      # Sample size
alpha = 0.05          # Significance level
num_simulations = 10000  # Number of simulations

# Critical value for Z test at alpha=0.05 (two-tailed test)
z_critical = norm.ppf(1 - alpha / 2)  # For two-tailed test

# Function to perform one hypothesis test
def perform_test(sample_mean, population_mean, population_std, sample_size):
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    p_value = 2 * (1 - norm.cdf(abs(z_score)))  # Two-tailed p-value
    return p_value < alpha, z_score  # True if rejected, False if not

# Simulate Type I error (H₀ is true)
type_i_errors = []
for _ in range(num_simulations):
    sample_data = np.random.normal(population_mean, population_std, sample_size)
    sample_mean = np.mean(sample_data)
    reject_null, z_score = perform_test(sample_mean, population_mean, population_std, sample_size)
    type_i_errors.append(reject_null)

# Simulate Type II error (H₁ is true)
alternative_mean = 55  # Assume the alternative hypothesis is true (µ = 55)
type_ii_errors = []
for _ in range(num_simulations):
    sample_data = np.random.normal(alternative_mean, population_std, sample_size)
    sample_mean = np.mean(sample_data)
    reject_null, z_score = perform_test(sample_mean, population_mean, population_std, sample_size)
    type_ii_errors.append(not reject_null)

# Calculate Type I and Type II errors
type_i_error_rate = np.mean(type_i_errors)
type_ii_error_rate = np.mean(type_ii_errors)

# Output the error rates
print(f"Type I Error Rate (α): {type_i_error_rate:.4f}")
print(f"Type II Error Rate (β): {type_ii_error_rate:.4f}")
print(f"Power of the test (1 - β): {1 - type_ii_error_rate:.4f}")

# Plot the distribution of Z-scores
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.plot(x, y, label='Standard Normal Distribution')
plt.axvline(x=z_critical, color='red', linestyle='--', label=f"Critical Z = ±{z_critical:.2f}")
plt.axvline(x=-z_critical, color='red', linestyle='--')
plt.title('Type I and Type II Errors in Hypothesis Testing')
plt.xlabel('Z-Score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Given values
P_A = 0.01  # Prior probability: Probability that a person has the disease
P_B_given_A = 0.95  # Likelihood: Probability of testing positive given the person has the disease
P_B_given_not_A = 0.05  # Likelihood: Probability of testing positive given the person does not have the disease
P_not_A = 0.99  # Prior probability: Probability that a person does not have the disease

# Step 1: Calculate P(B), the total probability of testing positive
P_B = (P_B_given_A * P_A) + (P_B_given_not_A * P_not_A)

# Step 2: Apply Bayes' Theorem to calculate P(A | B)
P_A_given_B = (P_B_given_A * P_A) / P_B

# Output the results
print(f"Probability of having the disease given a positive test result (P(A | B)): {P_A_given_B:.4f}")
```

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Simulate a contingency table (Observed Frequencies)
# Rows: Gender (Male, Female)
# Columns: Preference (Likes, Dislikes)

data = np.array([[30, 10],  # 30 Males like the product, 10 Males dislike the product
                 [25, 15]])  # 25 Females like the product, 15 Females dislike the product

# Create a DataFrame for better visualization
df = pd.DataFrame(data, columns=['Likes', 'Dislikes'], index=['Male', 'Female'])
print("Observed Frequencies:\n", df)

# Perform the Chi-Square Test for Independence
chi2_stat, p_val, dof, expected = chi2_contingency(data)

# Output the results
print("\nChi-Square Statistic:", chi2_stat)
print("P-Value:", p_val)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:\n", expected)

# Interpretation
alpha = 0.05  # Significance level
if p_val < alpha:
    print("\nReject the null hypothesis: The two variables are dependent.")
else:
    print("\nFail to reject the null hypothesis: The two variables are independent.")
```

```python
import numpy as np

# Observed data
observed = np.array([[30, 10],  # Male: 30 Likes, 10 Dislikes
                     [25, 15]])  # Female: 25 Likes, 15 Dislikes

# Step 1: Calculate the row totals, column totals, and grand total
row_totals = observed.sum(axis=1)  # Sum along rows (Male, Female)
col_totals = observed.sum(axis=0)  # Sum along columns (Likes, Dislikes)
grand_total = observed.sum()      # Sum of all entries

# Step 2: Calculate expected frequencies
expected = np.outer(row_totals, col_totals) / grand_total

# Output the expected frequencies
print("Observed Frequencies:")
print(observed)

print("\nExpected Frequencies:")
print(expected)
```
